Para crear un número complejo $z = a + jb$ (note que se usa $j$ para representar la unidad imaginaria) hacemos `z = complex(a, b)`

## Serie Compleja de Fourier para resolver EDOs

Nuestro objetivo es resolver la ecuación diferencial (encontrar la solución al estado estable para la carga $q(t)$) para un circuito RLC con una fuente de voltaje descrito por una función periódica $v(t)$:



La carga en el capacitor está modelada por la ecuación:
\begin{equation} L \frac{d^2q}{dt^2} + R\frac{dq}{dt} + \frac{1}{C} q = v(t). \label{eq:edo} \tag{1}\end{equation}

Podemos expandir $v(t)$ en su serie compleja de Fourier: $v(t) = \displaystyle\sum_{n = -\infty}^\infty c_n e^{i \omega_n t}$ y al usar el principio de superposición resolver
$$ L \frac{d^2q}{dt^2} + R\frac{dq}{dt} + \frac{1}{C} q = e^{i \omega_n t},$$

que podemos resolver usando coeficientes indeterminados:

- Proponemos $q_n(t) = A_n e^{i \omega_n t}$ y obtendremos que $A_n = \dfrac{C}{(1-LC\omega_n^2) + RC\omega_n i}$

Por simplicidad asumamos que $f(t)$ tiene período $T = 2 \pi$ y entonces:

$$A_n = \dfrac{C}{(1-LCn^2) + RCn i}.$$

La solución a la ecuación $\eqref{eq:edo}$ es entonces:

$$q(t) = \sum_{n = -\infty}^\infty c_n A_n e^{i \omega_n t} = \sum_{n = - \infty}^\infty c_n \dfrac{C}{(1-LCn^2) + RCn i} e^{i n t}.$$

```python
# valores de L, R, C
L, R, C = 1, 1, 1

def An(n):
    # denominador de An
    den = complex(1-L*C*n**2, R*C*n)
    return C / den

print(f"El coeficiente A3 = {An(3)}")
```

El método `quad` que hemos usado para integrar no puede manejar terminos complejos. Recuerde que si $f(t$ es compleja
$$\int f(t) dt = \int \operatorname{Re}(f(t)) dt + j \int \operatorname{Im}(f(t)) dt$$

```python
def f(t):
    if t > 0:
        return 1
    else: 
        return -1

def cn(n):
    j = complex(0, 1)
    
    # integramos la parte real
    integrando_real = lambda t: f(t)*np.exp(-j*n*t).real
    parte_real = quad(integrando_real, -np.pi, np.pi)[0]
    
    # integramos la parte imaginaria
    integrando_imaginaria = lambda t: f(t)*np.exp(-j*n*t).imag
    parte_imaginaria = quad(integrando_imaginaria, -np.pi, np.pi)[0]
    
    coef = (1/ (2*np.pi)) * complex(parte_real, parte_imaginaria)
    
    return coef
```

```python
cn(1)
```

Recuerde la relación entre los coeficientes $c_n$, $a_0, a_n$ y $b_n$:
- $a_0 = c_0$
- $a_n = c_n - c_{-n}$
- $b_n = i(c_n  - c_{-n})$

```python
# tamaño de la grafica
rcParams['figure.figsize'] = 6, 4

# Terminos de la serie
Nmax = 30


# calculamos la serie de la función de la entrada
a0 = cn(0).real

an = lambda n: 2*cn(n).real

bn = lambda n: -2*cn(n).imag

ans = [an(n) for n in range(1, Nmax+1)]
bns = [bn(n) for n in range(1, Nmax+1)]

v = lambda t: a0 + sum([ans[n-1]*np.cos(n*t) + bns[n-1]*np.sin(n*t) for n in range(1, Nmax+1)])

### Calculamos la solución a la edo
## recuerde que los coeficientes de la solución son los cn*An
a0_sol = (cn(0)*An(0)).real

an_sol = lambda n: 2 * (cn(n)*An(n)).real

bn_sol = lambda n: -2 * (cn(n)*An(n)).imag

ans_sol = [an_sol(n) for n in range(1, Nmax+1)]
bns_sol = [bn_sol(n) for n in range(1, Nmax+1)]


q = lambda t: a0_sol + sum([ans_sol[n-1]*np.cos(n*t) + bns_sol[n-1]*np.sin(n*t) for n in range(1, Nmax+1)])


t = np.linspace(0, 6*np.pi, 200)
plt.plot(t, v(t), label='$v(t)$', color='green')

plt.plot(t, q(t), label='$q(t)$', color='red')

plt.grid()
plt.legend()
plt.show()
```

```python
# Veamos como evoluciona la solución al variar el tiempo
from ipywidgets import interact, FloatSlider
from matplotlib import rcParams
rcParams['figure.figsize'] = 6, 4

def plot_sol(tmax):
    t = np.linspace(0, tmax, 200)
    plt.plot(t, q(t), label='$q(t)$', color='red')
    plt.plot(t, v(t), label='$v(t)$', color='green')
    plt.scatter(tmax, q(tmax), color='red')

    plt.ylim(-2, 2)
   
    plt.grid()
    plt.legend(loc='upper left')
    plt.show()
    
interact(plot_sol, tmax= FloatSlider(min = 0.01, max = 50, step=0.5));
```

Considere el siguiente sistema masa-resorte:



se aplica una fuerza externa periódica dada por $f(t) = \left| \sin t \right|$ con período $T = 2 \pi$.
- Encuentre la respuesta (solución al estado estable) de: $$ m \frac{d^2 x}{dt^2} + b \frac{dx}{dt} + k x = f(t)$$ usando superposición con $\operatorname{Nmax} = 30$.
 - $m$ es la mada del bloque
 - $b$ es la constante de amortigüamiento viscoso
 - $k$ es la constante del resorte


- Genere gráficas para $x(t), v(t)$ y $f(t)$.

## 1. Expansión en Serie
Expanda $f(t)$ como una serie compleja de Fourier $\hat{f}(t) = \displaystyle\sum_{n = -\infty}^\infty c_n e^{i\omega_n t}$.

```python
# Cantidad de terminos en la serie
Nmax = 30

# Defina el semiperiodo de la función
L = np.pi

# Defina la función f
# referencia de como la creo arriba (por valor absoluto)
def f(t):
    if t > 0:
        return np.sin(t)
    else: 
        return (-1)*np.sin(t)

# Defina una función que calcule los coeficientes cn para un n dado
# la copie de arribita
def calcular_cn(n):
    j = complex(0, 1)
    
    # integramos la parte real
    integrando_real = lambda t: f(t)*np.exp(-j*n*t).real
    parte_real = quad(integrando_real, -np.pi, np.pi)[0]
    
    # integramos la parte imaginaria
    integrando_imaginaria = lambda t: f(t)*np.exp(-j*n*t).imag
    parte_imaginaria = quad(integrando_imaginaria, -np.pi, np.pi)[0]
    
    coef = (1/ (2*np.pi)) * complex(parte_real, parte_imaginaria)
    
    return coef
```

```python
############ EN ESTA PARTE NO ES NECESARIO MODIFICAR NADA
# Convirtamos la serie 
a0 = calcular_cn(0).real

an = lambda n: 2 * calcular_cn(n).real
bn = lambda n: -2 * calcular_cn(n).imag

ans = [an(n) for n in range(1, Nmax+1)]
bns = [bn(n) for n in range(1, Nmax+1)]

f_serie = lambda t: a0 + sum([ans[n-1]*np.cos(n*np.pi*t / L) + bns[n-1]*np.sin(n*np.pi*t / L) for n in range(1, Nmax+1)])
```

Veamos como se ve la gráfica de su expansión en serie

```python
############ EN ESTA PARTE NO ES NECESARIO MODIFICAR NADA
# Tamaño de la gráfica
rcParams['figure.figsize'] = 6, 4

# Grafiquemos 3 períodos
t = np.linspace(0, 3*L, 200)

plt.plot(t, f_serie(t), color='green', label='$\hat{f}(t)$')

plt.legend()
plt.grid()
plt.show()
```

## 2. Superposición

- Proponga una solución $x_p = $
- Use coeficientes indeterminados para encontrar los coeficientes $An$ (en $x_p$)

## 2.1. Expansión en Serie

- $m\left ( \frac{d^2x}{dt}  \right )+ b \left ( \frac{dx}{dt}  \right ) + Kx = x(t)$-
- $m\left ( \frac{d^2x}{dt}  \right )+ b \left ( \frac{dx}{dt}  \right ) + Kx =  e^{iwt}$
- $x(t) = A_ne^{iwt}$
- $x'(t) = (A_n)ine^{iwt}$
- $x''(t) = -(A_n)n^2e^{iwt}$

```python
# Defina los valores de la masa y las constantes de amorigüamiento
m = 5
b = 5
k = 28

# Defina una función que le calcule los coeficientes An (que enconstró al usar coef. indeterminados)

def An(n):
    denominador = complex(k-m*n**2, b*n)
    return 1/denominador
```

## 3. Solución

Escribamos la solución a la ecuación usando superposición.

```python
############ EN ESTA PARTE NO ES NECESARIO MODIFICAR NADA
### Calculamos la solución a la edo
a0_sol = (cn(0)*An(0)).real

an_sol = lambda n: 2 * (cn(n)*An(n)).real

bn_sol = lambda n: -2 * (cn(n)*An(n)).imag

ans_sol = [an_sol(n) for n in range(1, Nmax+1)]
bns_sol = [bn_sol(n) for n in range(1, Nmax+1)]

x = lambda t: a0_sol + sum([ans_sol[n-1]*np.cos(n*np.pi*t /L ) + bns_sol[n-1]*np.sin(n*np.pi*t /L ) for n in range(1, Nmax+1)])

```

```python
############ EN ESTA PARTE NO ES NECESARIO MODIFICAR NADA
# Tamaño de la gráfica
rcParams['figure.figsize'] = 6, 4

# Grafiquemos 3 períodos
t = np.linspace(0, 5 * L, 200)

plt.plot(t, x(t), color='red', label='$x(t)$')

plt.legend()
plt.grid()
plt.show()
```
